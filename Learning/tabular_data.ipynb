# Scraping of Tabular data

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get("https://www.iplt20.com/auction/2025")
soup = BeautifulSoup(url.content,'lxml')
print(soup)
```

```python
table = soup.find("table", id="t3-delhi-capitals")
title = table.find_all("th")
print(title)
```

```python
header = []
for i in title:
    header.append(i.get_text())
print(header)
# Droping the last emtpy String
header = header[:-1]
print(f"After removing empty string: {header}")
```

```python
# Finding the Table data
data = table.find_all('td')
print(data)
```

```python
# Extracting the text from the  data
data_text = [tag.get_text() for tag in data]
print(data_text)
```

```python
# Applying .strip to the data_text
lst = []
for i in data_text:
    lst.append(i.strip())

print(len(lst))
for i in lst:
    print(i)
```

```python
# Removing Empty Strings
cleaned_lst = list(filter(None,lst))
for i in cleaned_lst:
    print(i)
```

```python
num_columns = 5
structured_data = [cleaned_lst[i:i + num_columns] for i in range(0, len(cleaned_lst), num_columns)]
print(structured_data)
```

```python
df = pd.DataFrame(structured_data,columns=header)
print(df)
```

```python
# Saving the file as csv with a different name to avoid PermissionError
df.to_csv("IplTABLE.csv", index=False)
```
