```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get("https://www.iplt20.com/auction/2025")
soup = BeautifulSoup(url.content,'lxml')
print(soup)
```

```python
table = soup.find("div", id ="autab3-2022")
print(table)
```

```python
# Finding the team names from headers
teams_name = table.find_all('h2')
teams = [team.get_text() for team in teams_name]
print(len(teams))
print(teams)
```

```python
# Finding the table Column names 
col_name = soup.find("table", id="t3-delhi-capitals")
title = col_name.find_all("th")
print(title)
```

```python
header = []
for i in title:
    header.append(i.get_text())
print(header)
# Droping the last emtpy String
header = header[:-1]
print(f"After removing empty string: {header}")
```

```python
# Finding the Table data
data = table.find_all('td')
print(len(data))
print(data)
```

```python
# Finding the actual Text from Table data
data_text = [tag.get_text() for tag in data]
print(len(data_text))
print(data_text)
```

```python
# Applying strip and filter so that data can be represented in structured way
# Step-1 Applying .strip method to reduce spaces
lst = []
for i in data_text:
    lst.append(i.strip())

# Applying .filter() to drop the empty strings
cleaned_lst = list(filter(None,lst))
for i in cleaned_lst:
    print(i)
```

```python
num_columns = 5
structured_data = [cleaned_lst[i:i + num_columns] for i in range(0, len(cleaned_lst), num_columns)]
print(structured_data)
```

```python
# Converting it in dataFrame
df = pd.DataFrame(structured_data,columns=header)
print(df)
```

```python
# Correcting the SR. NO Column
df['Sr. No.'] = df.index +1
print(df)
```

```python
# Adding one more column so that each player specified by thier team
df['Team'] = ""
df.loc[(df['Sr. No.'] >= 1) & (df['Sr. No.'] <= 20), 'Team'] = teams[0]
df.loc[(df['Sr. No.'] >= 21) & (df['Sr. No.'] <= 39), 'Team'] = teams[1]
df.loc[(df['Sr. No.'] >= 40) & (df['Sr. No.'] <= 59), 'Team'] = teams[2]
df.loc[(df['Sr. No.'] >= 60) & (df['Sr. No.'] <= 74), 'Team'] = teams[3]
df.loc[(df['Sr. No.'] >=75) & (df['Sr. No.'] <= 93), 'Team'] = teams[4]
df.loc[(df['Sr. No.'] >= 94) & (df['Sr. No.'] <=111), 'Team'] = teams[5]
df.loc[(df['Sr. No.'] >= 112) & (df['Sr. No.'] <= 134), 'Team'] = teams[6]
df.loc[(df['Sr. No.'] >= 135) & (df['Sr. No.'] <= 148), 'Team'] = teams[7]
df.loc[(df['Sr. No.'] >= 149) & (df['Sr. No.'] <= 167), 'Team'] = teams[8]
df.loc[(df['Sr. No.'] >= 168) & (df['Sr. No.'] <= 182), 'Team'] = teams[9]
```

```python
# After Creating Team Column now checking the entire dataFrame
print(df)
```

```python
# Saving the entire data frame in csv format
df.to_csv("IplAuctionRecord.csv", index=False)
```
